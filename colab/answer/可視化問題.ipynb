## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
df_scatter = pd.read_csv('../quizzes/matplotlib_quiz/scatter_data.csv')
fig2, ax2 = plt.subplots()
for name, group in df_scatter.groupby("class"):
    ax2.scatter(group["length (cm)"], group['width (cm)'], label = "class" + str(name))
ax2.grid(True)
ax2.set_title('クラス0～2のlengthとwidthに関する散布図')
ax2.set_xlabel('length（cm）')
ax2.set_ylabel('width（cm）')
ax2.legend()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
df_line = pd.read_csv(r'..\quizzes\matplotlib_quiz\line_data.csv')
fig3, ax3 = plt.subplots()
ax3.plot(df_line["期間"], df_line["件数"], marker = 'o')
ax3.grid(True)
ax3.set_title('期間ごとの件数の推移')
ax3.set_xlabel('期間')
ax3.set_ylabel('件数')
ax3.tick_params(axis='x', labelrotation=30)

fig4, ax4 = plt.subplots()
ax4.bar(df_line['期間'], df_line["費用"])
ax4.tick_params(axis='x', labelrotation=30)
ax4.grid(True)
ax4.set_title('期間ごとの費用の推移')
ax4.set_xlabel('期間')
ax4.set_ylabel('費用(円)')
```
